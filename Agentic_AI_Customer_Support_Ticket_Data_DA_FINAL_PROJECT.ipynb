#Dataset Information

Source:  Kaggle

Year / Timeline:  2025-2026

Dataset Description:

Ticket_ID – It represents the unique identification number assigned to each customer support ticket.

Customer_ID – It indicates the unique identification number assigned to each customer in the system.

Customer_Name – It tells us the name of the customer who raised the support request.

Customer_Email – It provides the email address of the customer used for communication.

Customer_Country – It indicates the country from which the customer is contacting support.

Customer_Language – It shows the preferred language used by the customer during communication.

Channel – It represents the medium through which the customer contacted support (such as email, chat, or phone).

Query_Category – It identifies the main category of the customer's issue or request.

Query_Subcategory – It provides a more detailed classification of the customer's problem.

Priority_Level – It indicates the urgency level of the support ticket (low, medium, or high).

Sentiment_Score – It represents the emotional tone of the customer's message, indicating whether the sentiment is positive, neutral, or negative.

Resolution_Status – It shows whether the issue has been resolved, pending, or still in progress.

Resolution_Time – It indicates the time taken by the support team or AI system to resolve the customer query.

Customer_Feedback_Rating – It represents the rating given by the customer after the issue is resolved.

Timestamp – It records the date and time when the customer support ticket was created.

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

palette = ["#E67E22", "#D35400", "#A04000", "#8B4513"]  # orange-brown theme
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Query_Category', data=df, color=primary_color)
plt.title("Distribution of Tickets by Query Category")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Channel', data=df, color=secondary_color)
plt.title("Customer Support Requests by Channel")
plt.show()
```

```python
channel_counts = df['Channel'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(channel_counts, labels=channel_counts.index, autopct='%1.1f%%', colors=palette)
plt.title("Customer Support Channels")
plt.show()
```
